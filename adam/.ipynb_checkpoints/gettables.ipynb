```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://fbref.com/en/squads/b8fd03ef/2018-2019/all_comps/Manchester-City-Stats-All-Competitions"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Alle Tabellen mit der Klasse 'stats_table' finden
tables = soup.findAll('table', {'class': 'stats_table'})

# DataFrames für jede gefundene Tabelle erstellen und in einer Liste speichern
dfs = []
for table in tables:
    dfs.append(pd.read_html(str(table))[0])

# Alle DataFrames zu einem einzigen DataFrame zusammenführen
final_df = pd.concat(dfs, ignore_index=True)

print(final_df)
```
