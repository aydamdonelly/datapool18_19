# Aggregating all player data for in-depth analyses and comparisons

```python
# libraries
import pandas as pd
import numpy as np
import time
import re
```

#### Getting the standard stats for all players playing for clubs from the Top5 leagues in 2018/19

```python
url_list = open('urls/top5_league_clubs.txt').read().splitlines()
print(url_list[0])
```

```python
from urllib.parse import urlparse
# empty list to store dataframes
dfs = []

# delay between requests (in seconds)
delay_between_requests = 5  

for url in url_list:
    try:
        tables = pd.read_html(url)

        df = tables[1]
        # drop the last two rows
        df = df.iloc[:-2]
        
        # extract club name from the URL using regex
        club_name_match = re.search(r"all_comps/(.*?)-Stats-All-Competitions", url)
        if club_name_match:
            club_name = club_name_match.group(1).replace('-', ' ')
        else:
            raise ValueError("Club name not found in URL pattern.")
        
        df['Club'] = club_name
        
        # append the dataframe to the list
        dfs.append(df)
        
        # delay before making the next request
        time.sleep(delay_between_requests)
    except Exception as e:
        print(f"Error reading data from {url}: {str(e)}")

# concatenate all dataframes into one
total_df = pd.concat(dfs, ignore_index=True)
```

```python
total_df.info()
```

```python
# dropping per90 stats and matches column
cols_to_drop = total_df.columns[23:34]

total_df.drop(columns=cols_to_drop, inplace=True)
total_df.head()
```

```python
total_df.columns
```

```python
total_df[('Unnamed: 4_level_0','MP')] = total_df[('Unnamed: 4_level_0','MP')].fillna(total_df[('Playing Time', 'MP')])
```

```python
total_df.info()
```

```python
total_df = total_df.iloc[:, :-1]
```

```python
df_clean_total = total_df.copy()
```

```python
df_clean_total.columns = df_clean_total.columns.get_level_values(1)
```

```python
df_clean_total.rename(columns={'': 'Club'}, inplace=True)
```

```python
df_clean_total
```

```python
df_clean_total.insert(1, "Club", df_clean_total.pop("Club"))
```

```python
# cleaning Nation values so only the uppercase string will be kept
df_clean_total['Nation'] = df_clean_total['Nation'].str.split().str[1]
```

```python
df_clean_total.head()
```

```python
# drop rows where the value for "Min" column equals zero or is NaN
df_clean_total = df_clean_total[(df_clean_total["Min"] != 0) & (~df_clean_total["Min"].isna())]
```

```python
df_clean_total.info()
```

```python
cols_to_parse = ["Age", "MP", "Min", "Gls", "Ast", "G+A", "G-PK", "PK", "PKatt", "CrdY", "CrdR", "PrgC", "PrgP", "PrgR"]

for col in cols_to_parse:
    # convert the float values in the column to integers
    df_clean_total[col] = df_clean_total[col].astype(int)
```

```python
df_clean_total.info()
```

```python
df_clean_total.to_csv("t5_leagues_players_standard.csv", index=False)
```
