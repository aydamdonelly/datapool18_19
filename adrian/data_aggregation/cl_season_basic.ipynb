```python
# creating list of urls for each top5 league club from the 18-19 season
url_list = open('urls/cl_matchday_reports.txt').read().splitlines()
print(url_list[0])
print(len(url_list))
```

```python
tables = pd.read_html(url_list[0], attrs={"id": "stats_822bd0ba_summary"})
df = tables[0]
df.head()
```

```python
df.info()
```

```python
from urllib.parse import urlparse
import time

# Initialize an empty list to store dataframes
dfs = []

# Set the delay between requests (in seconds)
delay_between_requests = 2  # Adjust this value as needed

for url in url_list:
    try:
        # Read HTML tables from the URL
        tables = pd.read_html(url, attrs={"id": "stats_822bd0ba_summary"})
        df = tables[0]
        
        # append df to the list
        dfs.append(df)
        
        # Introduce a delay before making the next request
        time.sleep(delay_between_requests)
    except Exception as e:
        print(f"Error reading data from {url}: {str(e)}")

# Concatenate all dataframes into one
total_df = pd.concat(dfs, ignore_index=True)
```

```python
df_clean_total = total_df.copy()
```

```python
df_clean_total.columns = df_clean_total.columns.get_level_values(1)
```

```python
df_clean_total.columns
```

```python
# cleaning Nation values so only the uppercase string will be kept
df_clean_total['Nation'] = df_clean_total['Nation'].str.split().str[1]
```

```python
df_clean_total.dropna(subset=['#'], inplace=True)
```

```python
df_clean_total["#"] = df_clean_total["#"].astype("int64")
```

```python
df_clean_total.info()
```

```python
df_clean_total.to_csv("cl_season_standard_stats.csv", index = False)
```
