```python
import pandas as pd

df = pd.read_csv("data/pl_club_results.csv")
```

```python
# Initialize new columns for accumulated goals and xG
df['Accumulated GF'] = 0
df['Accumulated xG'] = 0

# Initialize dictionary to store each club's current accumulated goals and xG
club_info = {}

# Loop to calculate accumulated goals and xG for each club
for index, row in df.iterrows():
    club = row['Club']
    gf = row['GF']
    xg = row['xG']
    
    # Initialize club info if not already initialized
    if club not in club_info:
        club_info[club] = {'Accumulated GF': 0, 'Accumulated xG': 0}
    
    # Update accumulated goals and xG
    club_info[club]['Accumulated GF'] += gf
    club_info[club]['Accumulated xG'] += xg
    
    # Update DataFrame with accumulated goals and xG
    df.at[index, 'Accumulated GF'] = club_info[club]['Accumulated GF']
    df.at[index, 'Accumulated xG'] = club_info[club]['Accumulated xG']

# Filter for matchweek 38
end_of_season_df = df[df['Round'] == "Matchweek 38"]

# Calculate the difference and average per matchday
end_of_season_df['xG_difference'] = end_of_season_df['Accumulated GF'] - end_of_season_df['Accumulated xG']
# lets just take the absolute numbers
# end_of_season_df['xG_diff_per_game'] = end_of_season_df['xG_difference'] / 38

end_of_season_df
```
